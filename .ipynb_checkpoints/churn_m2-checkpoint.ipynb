```python
import pickle
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, roc_auc_score

# حملي الـ data
X_train_sm = pickle.load(open('X_train_sm.pkl', 'rb'))
y_train_sm = pickle.load(open('y_train_sm.pkl', 'rb'))
X_test     = pickle.load(open('X_test.pkl', 'rb'))
y_test     = pickle.load(open('y_test.pkl', 'rb'))

print("✅ Data loaded!")
print("X_train_sm:", X_train_sm.shape)
print("X_test:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_sm, y_train_sm)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]

print("✅ Random Forest ROC-AUC:", round(roc_auc_score(y_test, y_prob_rf), 4))
print(classification_report(y_test, y_pred_rf))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=100, random_state=42, eval_metric='logloss')
xgb.fit(X_train_sm, y_train_sm)

y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]

print("✅ XGBoost ROC-AUC:", round(roc_auc_score(y_test, y_prob_xgb), 4))
print(classification_report(y_test, y_pred_xgb))
```

```python
results = {
    'Model':     ['Random Forest', 'XGBoost'],
    'ROC-AUC':   [round(roc_auc_score(y_test, y_prob_rf), 4),
                  round(roc_auc_score(y_test, y_prob_xgb), 4)],
}
print(pd.DataFrame(results))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb.fit(X_train_sm, y_train_sm)

y_pred_gb = gb.predict(X_test)
y_prob_gb = gb.predict_proba(X_test)[:, 1]

print("✅ Gradient Boosting ROC-AUC:", round(roc_auc_score(y_test, y_prob_gb), 4))
print(classification_report(y_test, y_pred_gb))
```

```python
comparison = pd.DataFrame({
    'Model':            ['Random Forest', 'XGBoost', 'Gradient Boosting'],
    'ROC-AUC':          [0.8169, 0.8079, 0.8321],
    'Accuracy':         [0.77, 0.76, 0.75],
    'Recall (Churn)':   [0.64, 0.64, 0.78],
    'Precision(Churn)': [0.56, 0.54, 0.52],
})
print(comparison.to_string(index=False))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
import pandas as pd
from sklearn.metrics import roc_auc_score

# load data
import pickle
X_train_sm = pickle.load(open('X_train_sm.pkl', 'rb'))
y_train_sm = pickle.load(open('y_train_sm.pkl', 'rb'))
X_test     = pickle.load(open('X_test.pkl', 'rb'))
y_test     = pickle.load(open('y_test.pkl', 'rb'))

print("✅ Ready!")
```

```python
## Hyperparameter Tuning — Gradient Boosting

from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators':   [100, 200],
    'max_depth':      [3, 5],
    'learning_rate':  [0.05, 0.1],
}

grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_sm, y_train_sm)

print("✅ Best Params:", grid.best_params_)
print("✅ Best ROC-AUC:", round(grid.best_score_, 4))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators':   [100, 200],
    'max_depth':      [3, 5],
    'learning_rate':  [0.05, 0.1],
}

grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_sm, y_train_sm)

print("✅ Best Params:", grid.best_params_)
print("✅ Best ROC-AUC:", round(grid.best_score_, 4))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Train best model
best_gb = GradientBoostingClassifier(
    learning_rate=0.1,
    max_depth=5,
    n_estimators=200,
    random_state=42
)
best_gb.fit(X_train_sm, y_train_sm)

# Test
y_pred_best = best_gb.predict(X_test)
y_prob_best = best_gb.predict_proba(X_test)[:, 1]

print("✅ ROC-AUC:", round(roc_auc_score(y_test, y_prob_best), 4))
print(classification_report(y_test, y_pred_best))
```

```python
best_gb2 = GradientBoostingClassifier(
    learning_rate=0.05,
    max_depth=3,
    n_estimators=200,
    random_state=42
)
best_gb2.fit(X_train_sm, y_train_sm)

y_pred2 = best_gb2.predict(X_test)
y_prob2 = best_gb2.predict_proba(X_test)[:, 1]

print("✅ ROC-AUC:", round(roc_auc_score(y_test, y_prob2), 4))
print(classification_report(y_test, y_pred2))
```

```python
import pickle

pickle.dump(best_gb2, open('best_model.pkl', 'wb'))
print("✅ Model saved!")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_best)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                               display_labels=['No Churn', 'Churn'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix — Gradient Boosting')
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(y_test, y_prob_best)
plt.title('ROC Curve — Gradient Boosting')
plt.plot([0,1], [0,1], 'k--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Feature Importance
feature_names = X_train_sm.columns if hasattr(X_train_sm, 'columns') else [f'f{i}' for i in range(X_train_sm.shape[1])]

importance = pd.Series(best_gb2.feature_importances_, index=feature_names)
importance = importance.sort_values(ascending=True).tail(10)

plt.figure(figsize=(10, 6))
importance.plot(kind='barh', color='#7c3aed')
plt.title('Top 10 Feature Importance — Gradient Boosting')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```
