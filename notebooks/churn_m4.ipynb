```python
import mlflow
import pickle
import mlflow.sklearn


best_gb2 = pickle.load(open('best_model.pkl', 'rb'))

mlflow.set_experiment("churn_prediction")

with mlflow.start_run(run_name="GradientBoosting_Tuned"):
    mlflow.log_param("model", "GradientBoosting")
    mlflow.log_param("learning_rate", 0.05)
    mlflow.log_param("max_depth", 3)
    mlflow.log_param("n_estimators", 200)
    
    mlflow.log_metric("roc_auc", 0.8334)
    mlflow.log_metric("recall", 0.78)
    mlflow.log_metric("precision", 0.52)
    mlflow.log_metric("accuracy", 0.75)
    
    mlflow.sklearn.log_model(best_gb2, "gradient_boosting_model")
    
    print("✅ Experiment logged!")
    print("Run ID:", mlflow.active_run().info.run_id)
```
